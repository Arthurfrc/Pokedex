```python
# Célula 1: imports e path base
from pathlib import Path

# ajusta aqui se seu notebook estiver num lugar diferente
base_dir = Path('../PokemonTeamBuilder/assets/pokemons')

pokemon_map = {}
```

```python
# Célula 2: varre as sub-pastas e gera o require()
for folder in sorted(base_dir.iterdir()):
    if not folder.is_dir():
        continue

     # busca todos os .gif (ou outro formato) dentro da pasta
    for file in folder.glob('*.gif'):
        stem = file.stem  # ex: 'nidoran-f' ou 'nidoran-m'
        # filtra só os que terminam em -f ou -m (caso queira garantir)
        if stem.endswith('-f') or stem.endswith('-m'):
            # monta a linha de require de acordo com seu padrão de import
            path = f"require('../../assets/pokemons/{folder.name}/{file.name}')"
            pokemon_map[stem] = path

# imprime no formato que você já usava
for name, req in pokemon_map.items():
    print(f"{name}: {req},")
```

```python
# Célula 2: varre as sub-pastas e gera o require()
for folder in sorted(base_dir.iterdir()):
    if not folder.is_dir():
        continue

    # folder.name é algo como "0001-bulbasaur"
    id_name = folder.name

    # extrai o slug depois do primeiro hífen e força minúsculas
    # ex: "0001-bulbasaur" → ["0001","bulbasaur"]
    parts = id_name.split('-', 1)
    if len(parts) != 2:
        continue
    slug = parts[1].lower()

    # imprime a linha com as aspas e a vírgula
    print(f"{slug}: require('../../assets/pokemons/{id_name}/{slug}.gif'),")
```
